```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
%matplotlib inline
image=cv2.imread('Valley-Taurus-Mountains-Turkey.jpg')
image_copy=np.copy(image)
mdimg=cv2.cvtColor(image_copy,cv2.COLOR_BGR2RGB)
plt.imshow(mdimg)
```

```python
#Detecting corners

gry=cv2.cvtColor(mdimg,cv2.COLOR_RGB2GRAY)
ndimg=np.float32(gry)
corners=cv2.cornerHarris(ndimg,2,3,0.04)
#now dilate the corner image obtained
dilate=cv2.dilate(corners,None)
plt.imshow(dilate)
```

```python
thresh=0.1*dilate.max()

corner_copy=np.copy(image_copy)

for j in range(0,dilate.shape[0]):
    for i in range(0,dilate.shape[1]):
        if dilate[j,i]>thresh:
            cv2.circle(corner_copy,(i,j),2,(0,255,0),1)
plt.imshow(corner_copy)
```
